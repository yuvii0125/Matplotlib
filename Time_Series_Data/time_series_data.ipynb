```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
from datetime import datetime
```

```python
plt.figure(figsize=(40,50))  
```

```python
plt.style.use('seaborn-v0_8')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot(dates,y,marker='o')
date_format = mpl_dates.DateFormatter('%b , %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('../Datasets/time_series.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date',inplace=True)
price_data = df['Date']
price_close = df['Close']
```

```python
plt.plot(price_data,price_close,marker='o')
plt.tight_layout()
plt.show()
```
