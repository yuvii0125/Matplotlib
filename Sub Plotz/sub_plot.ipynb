```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv(r'../subplot_data.csv')
df.head()
```

```python
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']
```

```python
plt.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
plt.plot(ages,py_salaries,label='Python')
plt.plot(ages,js_salaries,label='JavaScript')

plt.legend()
plt.title('Median Salaries by Age(USD)')
plt.xlabel('Age')
plt.ylabel('Salaries in (USD)')

plt.tight_layout()
plt.show()
```

We will do it using axes 

```python
fig,ax = plt.subplots()

ax.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
ax.plot(ages,py_salaries,label='Python')
ax.plot(ages,js_salaries,label='JavaScript')

ax.legend()
ax.set_title('Median Salaries by Age(USD)')
ax.set_xlabel('Age')
ax.set_ylabel('Salaries in (USD)')

plt.tight_layout()
plt.show()
```

#### with ax we can make a multiple plot in 1 plot

#### nrows used for number of rows and ncols for number of cols

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
ax2.plot(ages,py_salaries,label='Python')
ax2.plot(ages,js_salaries,label='JavaScript')

ax1.legend()
ax1.set_title('Median Salaries by Age(USD)')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salaries in (USD)')

ax2.legend()
ax2.set_title('Median Salaries by Age(USD)')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries in (USD)')


plt.tight_layout()
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
ax2.plot(ages,py_salaries,label='Python')
ax2.plot(ages,js_salaries,label='JavaScript')

ax1.legend()
ax1.set_title('Median Salaries by Age(USD)')
ax1.set_ylabel('Salaries in (USD)')

ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries in (USD)')


plt.tight_layout()
plt.show()
```

we use sharex for same x axis and sharey for same y axis values in number line

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True)

ax1.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
ax2.plot(ages,py_salaries,label='Python')
ax2.plot(ages,js_salaries,label='JavaScript')

ax1.legend()
ax1.set_title('Median Salaries by Age(USD)')
ax1.set_ylabel('Salaries in (USD)')


ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries in (USD)')

plt.tight_layout()
plt.show()
```

```python
fig1,ax1 = plt.subplots()
fig2,ax2 = plt.subplots()

ax1.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All devs')
ax2.plot(ages,py_salaries,label='Python')
ax2.plot(ages,js_salaries,label='JavaScript')

ax1.legend()
ax1.set_title('Median Salaries by Age(USD)')
ax1.set_ylabel('Salaries in (USD)')


ax2.set_xlabel('Age')
ax2.set_ylabel('Salaries in (USD)')

plt.tight_layout()
plt.show()
```
